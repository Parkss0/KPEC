```python
from scipy.interpolate import make_interp_spline
import numpy as np
import pandas as pd

df['접수일'] = pd.to_datetime(df['접수일'])
df['YearMonth'] = df['접수일'].dt.to_period('M')
year_month_counts = df['YearMonth'].value_counts().sort_index()

# ✅ 시작~끝까지 빠진 월 없이 완전한 등간격 인덱스 생성
full_index = pd.period_range(
    start=year_month_counts.index.min(),
    end=year_month_counts.index.max(),
    freq='M'
)

# ✅ 빠진 월은 0으로 채움
year_month_counts = year_month_counts.reindex(full_index, fill_value=0)
year_month_labels = [str(x) for x in year_month_counts.index]

fig, ax = plt.subplots(figsize=(18, 7))

# 바 차트
x_numeric = np.arange(len(year_month_counts))
ax.bar(x_numeric, year_month_counts.values,
       color='steelblue', alpha=0.4, width=0.6, label='Monthly Count')

# 숫자 레이블 (0 제외)
for i, v in enumerate(year_month_counts.values):
    if v > 0:
        ax.annotate(f'{v}', (i, v), ha='center', va='bottom',
                    xytext=(0, 4), textcoords='offset points', fontsize=8)

# 스플라인 곡선
y_values = year_month_counts.values.astype(float)
x_smooth = np.linspace(x_numeric.min(), x_numeric.max(), 500)
spline = make_interp_spline(x_numeric, y_values, k=3)
y_smooth = spline(x_smooth)
y_smooth = np.clip(y_smooth, 0, None)  # 음수 방지

ax.plot(x_smooth, y_smooth, color='crimson', linewidth=2.5, label='Trend (Spline)')
ax.scatter(x_numeric, y_values, color='crimson', zorder=5, s=30)

# x축 레이블: 너무 많으면 3개월 간격으로 표시
tick_step = 3
ax.set_xticks(x_numeric[::tick_step])
ax.set_xticklabels(year_month_labels[::tick_step], rotation=60, ha='right', fontsize=10)

ax.set_title('Distribution of Applications by Year and Month (Equal Interval)', fontsize=16)
ax.set_xlabel('Year-Month', fontsize=13)
ax.set_ylabel('Number of Applications', fontsize=13)
ax.legend(fontsize=12)
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

### Treemap of Target Pathogen Type Distribution

To visualize the distribution of specific target pathogen types (e.g., 'SARS-CoV-2', 'Influenza A virus') from the `타깃병원체종류` (target pathogen type) column, we will extract the last component of the hierarchical string. A treemap will be used to effectively display the proportional distribution of these labels, ensuring a publication-quality output with clear labels and colors.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Function to extract the last label from the hierarchical string
def extract_last_label(pathogen_type):
    if pd.isna(pathogen_type):
        return None
    parts = str(pathogen_type).split('>')
    return parts[-1].strip() if parts else None

# Apply the function to create a new 'Last_Pathogen_Label' column
df['Last_Pathogen_Label'] = df['타깃병원체종류'].apply(extract_last_label)

# Get the value counts for these labels, dropping None values
label_counts = df['Last_Pathogen_Label'].dropna().value_counts()

# Filter out labels with very small counts to group them into 'Other'
# This helps in creating a cleaner chart for publication
threshold_percent = 0.02 # Labels less than 2% will be grouped into 'Other'

# Calculate percentages
percentages = label_counts / label_counts.sum()

# Separate labels that meet the threshold and those that don't
main_labels = percentages[percentages >= threshold_percent]
other_labels = percentages[percentages < threshold_percent]

# Combine small labels into 'Other'
if not other_labels.empty:
    main_labels['기타'] = other_labels.sum()

# Convert back to counts and sort for display
main_counts = (main_labels * label_counts.sum()).astype(int)
main_counts = main_counts.sort_values(ascending=False)

# Create the horizontal bar chart
plt.figure(figsize=(12, 8)) # Adjust figure size for better readability
ax = sns.barplot(x=main_counts.values, y=main_counts.index, palette='viridis', hue=main_counts.index, legend=False)

# Add title and labels with improved font sizes
plt.title('Distribution of Target Pathogen Types (Horizontal Bar Chart)', fontsize=16, pad=20)
plt.xlabel('Number of Applications', fontsize=14)
plt.ylabel('Pathogen Type', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add numerical labels on the bars
for p in ax.patches:
    width = p.get_width()
    plt.text(width + 0.5, # Position the text a bit to the right of the bar
             p.get_y() + p.get_height() / 2, # Position vertically centered
             '{:1.0f}'.format(width), # Format the count as integer
             ha='left', va='center', fontsize=10)

plt.grid(axis='x', linestyle='--', alpha=0.7) # Add a horizontal grid for better readability
plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()
```

### Treemap of Application Evaluation Type Distribution

To visualize the distribution of specific application evaluation types (e.g., '세포실험', '마우스', 'PK 채혈 및 분석') from the `신청평가유형` (Application Evaluation Type) column, we will extract the last component of the hierarchical string. A treemap will be used to effectively display the proportional distribution of these labels, ensuring a publication-quality output with clear labels and colors.

```python
import matplotlib.pyplot as plt
import squarify # For treemaps
import pandas as pd
import matplotlib.font_manager as fm

# Ensure Korean font is configured for new plots
# If not already run, this block ensures the font is set.
# Assuming NanumBarunGothic is installed from previous steps.
if 'NanumBarunGothic' not in plt.rcParams['font.family']:
    try:
        fm.fontManager.addfont('/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf')
        plt.rcParams['font.family'] = 'NanumBarunGothic'
        plt.rcParams['axes.unicode_minus'] = False
        print("Korean font re-configured for plotting.")
    except Exception as e:
        print(f"Could not configure Korean font: {e}")

# Function to extract the last label from the hierarchical string
def extract_last_evaluation_label(evaluation_type):
    if pd.isna(evaluation_type):
        return None
    parts = str(evaluation_type).split('>')
    return parts[-1].strip() if parts else None

# Apply the function to create a new 'Last_Evaluation_Label' column
df['Last_Evaluation_Label'] = df['신청평가유형'].apply(extract_last_evaluation_label)

# Get the value counts for these labels, dropping None values
label_counts = df['Last_Evaluation_Label'].dropna().value_counts()

# Filter out labels with very small counts to group them into '기타' (Other)
threshold_percent = 0.02 # Labels less than 2% will be grouped into 'Other'

# Calculate percentages
percentages = label_counts / label_counts.sum()

# Separate labels that meet the threshold and those that don't
main_labels_series = percentages[percentages >= threshold_percent]
other_labels_series = percentages[percentages < threshold_percent]

# Combine small labels into '기타' if there are any 'other' labels
if not other_labels_series.empty:
    main_labels_series = pd.concat([main_labels_series, pd.Series({'기타': other_labels_series.sum()})])

# Convert back to counts and sort for display in the treemap (largest first)
main_counts = (main_labels_series * label_counts.sum()).astype(int)
main_counts = main_counts.sort_values(ascending=False)

# Prepare data for treemap
sizes = main_counts.values
labels = [f'{idx}\n({p:.1f}%)' for idx, p in zip(main_counts.index, main_counts / main_counts.sum() * 100)]

# Generate a color palette (using a different one, e.g., 'Paired')
# You can try other palettes like 'tab10', 'Pastel1', 'Set3', 'Dark2', etc.
colors = [plt.cm.Paired(i/float(len(sizes))) for i in range(len(sizes))]

# Create the treemap
plt.figure(figsize=(15, 10)) # Adjust figure size for better readability
squarify.plot(sizes=sizes,
              label=labels,
              color=colors,
              alpha=0.8,
              text_kwargs={'fontsize': 10, 'color': 'black', 'fontweight': 'bold'}) # Changed text color to black for better contrast

plt.title('Distribution of Application Evaluation Types', fontsize=18, pad=20)
plt.axis('off') # Remove axes for a cleaner treemap look
plt.tight_layout()
plt.show()
```

```python
import re
import matplotlib.pyplot as plt
import seaborn as sns

# Function to extract the 2-character region after 5 digits
def extract_region(address):
    if pd.isna(address):
        return None
    match = re.search(r'\d{5}\s*([가-힣]{2})', str(address))
    if match:
        return match.group(1)
    return None

# Apply the function to the '주소' column to create a new '지역' column
df['지역'] = df['주소'].apply(extract_region)

# Get the value counts for the extracted regions, dropping None values
region_counts = df['지역'].dropna().value_counts()

# Filter out regions with very small counts to group them into 'Other'
threshold = 0.02 # Regions less than 2% will be grouped into 'Other'
filtered_counts = region_counts[region_counts / region_counts.sum() > threshold]
other_count = region_counts[region_counts / region_counts.sum() <= threshold].sum()

if other_count > 0:
    filtered_counts['기타'] = other_count

# Ensure the data is in ascending order for better visual (largest slices first)
# or keep it as is, depending on preference. For pie chart, largest first is common.
filtered_counts = filtered_counts.sort_values(ascending=False)

# Create the pie chart
fig, ax = plt.subplots(figsize=(10, 10)) # Adjust figure size as needed

# Define colors - using a palette for diversity
colors = sns.color_palette('pastel')[0:len(filtered_counts)]

# Explode the largest slice for emphasis (optional)
explode = [0.05 if i == 0 else 0 for i in range(len(filtered_counts))]

wedges, texts, autotexts = ax.pie(filtered_counts.values,
                                  labels=filtered_counts.index,
                                  autopct=lambda p: '{:.1f}%'.format(p) if p > 1 else '', # Only show percentages > 1%
                                  pctdistance=0.85,
                                  colors=colors,
                                  explode=explode,
                                  shadow=False, # Set to True for 3D effect
                                  startangle=90,
                                  textprops={'fontsize': 12})

# Add a title
ax.set_title('Distribution of Applications by Region', fontsize=16, pad=20)

# Equal aspect ratio ensures that pie is drawn as a circle.
ax.axis('equal')

# Adjust autopct text properties for better readability
for autotext in autotexts:
    autotext.set_color('black')
    autotext.set_fontsize(10)

plt.tight_layout()
plt.show()
```

### 셀 3: 전처리

선택된 컬럼들에 대해 요구사항에 맞춰 데이터를 정제하고 새로운 피처를 생성합니다. 결측값도 처리합니다.

```python
# 1. 텍스트 레벨 추출 함수 정의: '>' 기준으로 split 후 마지막 레벨 사용
def extract_last_level(text):
    if pd.isna(text):
        return None
    parts = str(text).split('>')
    return parts[-1].strip() if parts else None

# '타깃병원체종류', '시험종류', '신청평가유형' 컬럼에 함수 적용
df_ml['타깃병원체종류'] = df_ml['타깃병원체종류'].apply(extract_last_level)
df_ml['시험종류'] = df_ml['시험종류'].apply(extract_last_level)
df_ml['신청평가유형'] = df_ml['신청평가유형'].apply(extract_last_level)

# 2. '주소' 컬럼에서 시/도 추출
def extract_city(address):
    if pd.isna(address):
        return None
    # 우편번호(5자리 숫자) 제거 후 첫 번째 한글 단어 추출 시도
    match = re.search(r'\d{5}\s*([가-힣]+)', str(address))
    if match:
        return match.group(1)
    return None

df_ml['지역'] = df_ml['주소'].apply(extract_city)

# 3. '접수일' 컬럼 전처리: datetime 변환 및 파생 피처 생성 (연도, 월, 분기, Lag 피처)
df_ml['접수일'] = pd.to_datetime(df_ml['접수일'], errors='coerce')

# 날짜 변환 실패로 인해 생긴 NaN 행 제거 (선택적, 여기서는 유효한 데이터만 사용)
df_ml.dropna(subset=['접수일'], inplace=True)

df_ml['연도'] = df_ml['접수일'].dt.year
df_ml['월'] = df_ml['접수일'].dt.month
df_ml['분기'] = df_ml['접수일'].dt.quarter

# 'lag_1' 피처 생성: 직전 달에 같은 '신청평가유형'이 몇 건이었는지
# 먼저 'YearMonth' 컬럼 생성
df_ml['YearMonth'] = df_ml['접수일'].dt.to_period('M')

# 월별, 평가유형별 건수 집계
monthly_counts = df_ml.groupby(['YearMonth', '신청평가유형']).size().unstack(fill_value=0)

# 'lag_1' 피처 계산을 위해 정렬 후 shift
# 모든 가능한 YearMonth를 포함하여 완전한 시계열 생성
min_ym = df_ml['YearMonth'].min()
max_ym = df_ml['YearMonth'].max()
full_ym_range = pd.period_range(start=min_ym, end=max_ym, freq='M')

monthly_counts_full = monthly_counts.reindex(full_ym_range, fill_value=0)

# Lag feature 계산
lag_features = monthly_counts_full.shift(1).stack().reset_index(name='lag_1')
lag_features.rename(columns={'level_0': 'YearMonth', 'level_1': '신청평가유형'}, inplace=True)

# 기존 df_ml에 'lag_1' 병합
df_ml = pd.merge(df_ml, lag_features, on=['YearMonth', '신청평가유형'], how='left')

# 'lag_1' 결측값은 0으로 대체 (직전 달에 해당 평가유형 신청이 없었음)
df_ml['lag_1'].fillna(0, inplace=True)

# 더 이상 필요 없는 원본 '접수일', '주소', 'YearMonth' 컬럼 제거
df_ml.drop(columns=['접수일', '주소', 'YearMonth'], inplace=True)

# 4. 결측값 처리: 모든 컬럼의 결측값을 최빈값으로 대체
for col in df_ml.columns:
    if df_ml[col].isnull().any():
        most_frequent = df_ml[col].mode()[0]
        df_ml[col].fillna(most_frequent, inplace=True)

df_ml.head()
```

### 셀 4: 피처(X)와 타깃(y) 분리 + Label Encoding + 훈련/테스트 데이터 분리

데이터를 피처와 타깃으로 나누고, 범주형 데이터를 숫자로 인코딩한 후, 모델 훈련을 위한 데이터셋과 모델 평가를 위한 데이터셋으로 분리합니다.

```python
# 피처(X)와 타깃(y) 분리
X = df_ml.drop('신청평가유형', axis=1)
y = df_ml['신청평가유형']

# Label Encoding: 타깃 변수 (y) 인코딩
le_y = LabelEncoder()
y_encoded = le_y.fit_transform(y)

# Label Encoding: 피처 변수 (X) 중 범주형 컬럼 인코딩
# 인코딩할 컬럼 리스트
categorical_features = [
    '신청기관구분', '타깃병원체종류', '시험종류', '지역'
]

# 각 범주형 피처에 LabelEncoder 적용
le_encoders = {}
for col in categorical_features:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
    le_encoders[col] = le # 나중에 역변환을 위해 저장

# 훈련/테스트 데이터 분리 (8:2 비율, random_state 고정)
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

print("훈련 데이터셋 크기:", X_train.shape, y_train.shape)
print("테스트 데이터셋 크기:", X_test.shape, y_test.shape)
```

### 셀 5: RandomForest 모델 학습

Random Forest 분류 모델을 초기화하고 훈련 데이터로 학습시킵니다.

```python
# Random Forest 분류 모델 초기화 (n_estimators=100, random_state=42)
model = RandomForestClassifier(n_estimators=100, random_state=42)

# 모델 학습
model.fit(X_train, y_train)

print("Random Forest 모델 학습 완료!")
```

### 셀 6: 모델 평가

학습된 모델을 테스트 데이터에 적용하여 정확도(accuracy)와 상세 분류 리포트를 출력합니다.

```python
# 테스트 데이터로 예측 수행
y_pred = model.predict(X_test)

# 정확도 출력
accuracy = accuracy_score(y_test, y_pred)
print(f"모델 정확도 (Accuracy): {accuracy:.4f}")

# Classification Report 출력 (각 클래스별 성능)
# y_test에 실제로 존재하는 클래스만 추출하여 target_names를 생성
unique_test_labels = np.unique(y_test)
actual_target_names = le_y.inverse_transform(unique_test_labels)

print("\nClassification Report:")
print(classification_report(y_test, y_pred, labels=unique_test_labels, target_names=actual_target_names))
```

### 범주형 변수 간 연관성 분석: Cramér's V 및 카이제곱 검정

모델의 성능이 낮았던 점을 고려하여, 피처들 간의 연관성을 심도 있게 분석하여 데이터의 특성을 더 잘 이해하고 향후 모델 개선에 활용할 수 있습니다. 여기서는 범주형 변수 간의 연관성을 측정하는 `Cramér's V`와 통계적 유의미성을 검증하는 `카이제곱 검정`을 수행하고, 결과를 히트맵으로 시각화합니다.

**Cramér's V:** 두 명목 변수 간의 연관성 강도를 0과 1 사이의 값으로 나타냅니다 (0: 연관성 없음, 1: 완전한 연관성).

**카이제곱 검정 (Chi-square Test):** 두 변수가 서로 독립적인지 여부를 통계적으로 검증합니다. p-값이 낮을수록 두 변수 간에 유의미한 연관성이 있을 가능성이 높습니다 (일반적으로 p < 0.05).

이 분석에서는 이전에 라벨 인코딩된 피처(`X`)와 타깃 변수(`y_encoded`)를 사용합니다.

```python
# 필요한 라이브러리 import
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# Cramér's V 계산 함수
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    # 분모가 0이 되는 것을 방지하기 위해 max(0, ...) 처리
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    # 분모가 0이 되는 것을 방지하고, 값이 NaN이 되지 않도록 처리
    if min((kcorr-1), (rcorr-1)) == 0:
        return 0.0
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

print("Cramér's V 함수가 정의되었습니다.")
```

#### 데이터 준비

`X` 피처와 `y_encoded` 타깃 변수를 하나의 DataFrame으로 통합합니다. 이 DataFrame의 모든 컬럼은 연관성 분석을 위해 범주형으로 간주됩니다.

```python
# 피처(X)와 인코딩된 타깃(y_encoded)을 하나의 DataFrame으로 통합
df_association = X.copy()
df_association['신청평가유형'] = y_encoded # y_encoded에 해당하는 컬럼 이름 지정

# 연관성 분석에 사용할 컬럼 목록 (모든 컬럼이 범주형으로 간주됨)
categorical_cols = df_association.columns

print("연관성 분석을 위한 데이터프레임이 준비되었습니다.")
display(df_association.head())
```

#### Cramér's V 계산 및 히트맵 시각화

각 범주형 변수 쌍에 대한 Cramér's V 값을 계산하고 히트맵으로 연관성 강도를 시각화합니다.

```python
# Cramér's V 행렬 초기화
cramers_v_matrix = pd.DataFrame(index=categorical_cols, columns=categorical_cols, dtype=float)

# 모든 범주형 컬럼 쌍에 대해 Cramér's V 계산
for col1 in categorical_cols:
    for col2 in categorical_cols:
        if col1 == col2:
            cramers_v_matrix.loc[col1, col2] = 1.0 # 자기 자신과의 연관성은 1
        else:
            v_score = cramers_v(df_association[col1], df_association[col2])
            cramers_v_matrix.loc[col1, col2] = v_score

print("Cramér's V 연관성 행렬:")
display(cramers_v_matrix)

# Cramér's V 히트맵 시각화
plt.figure(figsize=(12, 10))
sns.heatmap(cramers_v_matrix, annot=True, cmap='viridis', fmt=".2f", linewidths=.5)
plt.title("Cramér's V Association Matrix", fontsize=16)
plt.xlabel("Features", fontsize=14)
plt.ylabel("Features", fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(rotation=0, fontsize=12)
plt.tight_layout()
plt.show()
```

#### 카이제곱 검정 (Chi-square Test) p-value 계산 및 히트맵 시각화

각 범주형 변수 쌍에 대해 카이제곱 검정의 p-value를 계산하고, 이를 히트맵으로 시각화하여 통계적 유의미성을 파악합니다. p-value가 낮을수록 (예: 0.05 미만) 두 변수 간의 연관성이 통계적으로 유의미하다고 볼 수 있습니다.

```python
# Chi-square p-value 행렬 초기화
chi2_p_value_matrix = pd.DataFrame(index=categorical_cols, columns=categorical_cols, dtype=float)

# 모든 범주형 컬럼 쌍에 대해 Chi-square p-value 계산
for col1 in categorical_cols:
    for col2 in categorical_cols:
        if col1 == col2:
            chi2_p_value_matrix.loc[col1, col2] = 0.0 # 자기 자신과의 비교는 p-value 0 (완전 종속)
        else:
            # 교차표 생성
            cross_tab = pd.crosstab(df_association[col1], df_association[col2])
            # 모든 셀의 합이 0인 경우 (교차표가 비어있는 경우) 오류 방지
            if cross_tab.sum().sum() == 0:
                p_value = 1.0 # 연관성을 판단할 수 없으므로 독립으로 간주
            else:
                # 카이제곱 검정 수행
                # 최소 기대 빈도 5 이하인 셀이 많을 경우 결과가 부정확할 수 있음
                chi2, p_value, _, _ = chi2_contingency(cross_tab)
            chi2_p_value_matrix.loc[col1, col2] = p_value

print("Chi-square p-value 행렬:")
display(chi2_p_value_matrix)

# Chi-square p-value 히트맵 시각화
plt.figure(figsize=(12, 10))
# p-value가 0.05보다 작으면 유의미하다고 간주 (색상 반전)
sns.heatmap(chi2_p_value_matrix, annot=True, cmap='RdYlGn_r', fmt=".3f", linewidths=.5,
            cbar_kws={'label': 'p-value (Lower is more significant)'})
plt.title("Chi-square Test p-value Matrix", fontsize=16)
plt.xlabel("Features", fontsize=14)
plt.ylabel("Features", fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(rotation=0, fontsize=12)
plt.tight_layout()
plt.show()
```

모델이 어떤 클래스를 올바르게 예측하고 어떤 클래스를 혼동하는지 시각적으로 확인합니다. 이는 모델의 강점과 약점을 파악하는 데 도움을 줍니다.

### 셀 9: AUROC (ROC Curve) 시각화

다중 클래스 분류에서는 일반적으로 각 클래스에 대해 '일대다(One-vs-Rest)' 방식으로 ROC 곡선과 AUROC 값을 계산합니다. 이는 특정 클래스를 다른 모든 클래스와 구분하는 모델의 능력을 평가하는 데 유용합니다.

```python
# 필요한 라이브러리 import
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import LabelBinarizer
import numpy as np
import matplotlib.pyplot as plt

# y_test를 이진화 (One-vs-Rest 전략을 위해)
lb = LabelBinarizer()
y_test_binarized = lb.fit_transform(y_test)

# 각 클래스에 대한 예측 확률 얻기
y_score = model.predict_proba(X_test)

# 각 클래스에 대한 ROC 곡선 및 AUC 계산
n_classes = y_test_binarized.shape[1]
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_binarized[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# ROC 곡선 시각화
plt.figure(figsize=(10, 8))
# DeprecationWarning을 해결하기 위해 plt.cm.get_cmap 대신 matplotlib.colormaps 사용
colors = plt.colormaps['tab10'](np.linspace(0, 1, n_classes)) # 각 클래스별 색상

for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label=f'ROC curve of class {le_y.inverse_transform([i])[0]} (area = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2, label='Random classifier (area = 0.50)')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=14)
plt.ylabel('True Positive Rate', fontsize=14)
plt.title('Receiver Operating Characteristic (ROC) Curve for Multi-class', fontsize=16)
plt.legend(loc='lower right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
