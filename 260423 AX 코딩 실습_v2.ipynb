```python
from scipy.interpolate import make_interp_spline
import numpy as np
import pandas as pd

df['접수일'] = pd.to_datetime(df['접수일'])
df['YearMonth'] = df['접수일'].dt.to_period('M')
year_month_counts = df['YearMonth'].value_counts().sort_index()

# ✅ 시작~끝까지 빠진 월 없이 완전한 등간격 인덱스 생성
full_index = pd.period_range(
    start=year_month_counts.index.min(),
    end=year_month_counts.index.max(),
    freq='M'
)

# ✅ 빠진 월은 0으로 채움
year_month_counts = year_month_counts.reindex(full_index, fill_value=0)
year_month_labels = [str(x) for x in year_month_counts.index]

fig, ax = plt.subplots(figsize=(18, 7))

# 바 차트
x_numeric = np.arange(len(year_month_counts))
ax.bar(x_numeric, year_month_counts.values,
       color='steelblue', alpha=0.4, width=0.6, label='Monthly Count')

# 숫자 레이블 (0 제외)
for i, v in enumerate(year_month_counts.values):
    if v > 0:
        ax.annotate(f'{v}', (i, v), ha='center', va='bottom',
                    xytext=(0, 4), textcoords='offset points', fontsize=8)

# 스플라인 곡선
y_values = year_month_counts.values.astype(float)
x_smooth = np.linspace(x_numeric.min(), x_numeric.max(), 500)
spline = make_interp_spline(x_numeric, y_values, k=3)
y_smooth = spline(x_smooth)
y_smooth = np.clip(y_smooth, 0, None)  # 음수 방지

ax.plot(x_smooth, y_smooth, color='crimson', linewidth=2.5, label='Trend (Spline)')
ax.scatter(x_numeric, y_values, color='crimson', zorder=5, s=30)

# x축 레이블: 너무 많으면 3개월 간격으로 표시
tick_step = 3
ax.set_xticks(x_numeric[::tick_step])
ax.set_xticklabels(year_month_labels[::tick_step], rotation=60, ha='right', fontsize=10)

ax.set_title('Distribution of Applications by Year and Month (Equal Interval)', fontsize=16)
ax.set_xlabel('Year-Month', fontsize=13)
ax.set_ylabel('Number of Applications', fontsize=13)
ax.legend(fontsize=12)
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Function to extract the last label from the hierarchical string
def extract_last_label(pathogen_type):
    if pd.isna(pathogen_type):
        return None
    parts = str(pathogen_type).split('>')
    return parts[-1].strip() if parts else None

# Apply the function to create a new 'Last_Pathogen_Label' column
df['Last_Pathogen_Label'] = df['타깃병원체종류'].apply(extract_last_label)

# Get the value counts for these labels, dropping None values
label_counts = df['Last_Pathogen_Label'].dropna().value_counts()

# Filter out labels with very small counts to group them into 'Other'
# This helps in creating a cleaner chart for publication
threshold_percent = 0.02 # Labels less than 2% will be grouped into 'Other'

# Calculate percentages
percentages = label_counts / label_counts.sum()

# Separate labels that meet the threshold and those that don't
main_labels = percentages[percentages >= threshold_percent]
other_labels = percentages[percentages < threshold_percent]

# Combine small labels into 'Other'
if not other_labels.empty:
    main_labels['기타'] = other_labels.sum()

# Convert back to counts and sort for display
main_counts = (main_labels * label_counts.sum()).astype(int)
main_counts = main_counts.sort_values(ascending=False)

# Create the horizontal bar chart
plt.figure(figsize=(12, 8)) # Adjust figure size for better readability
ax = sns.barplot(x=main_counts.values, y=main_counts.index, palette='viridis', hue=main_counts.index, legend=False)

# Add title and labels with improved font sizes
plt.title('Distribution of Target Pathogen Types (Horizontal Bar Chart)', fontsize=16, pad=20)
plt.xlabel('Number of Applications', fontsize=14)
plt.ylabel('Pathogen Type', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add numerical labels on the bars
for p in ax.patches:
    width = p.get_width()
    plt.text(width + 0.5, # Position the text a bit to the right of the bar
             p.get_y() + p.get_height() / 2, # Position vertically centered
             '{:1.0f}'.format(width), # Format the count as integer
             ha='left', va='center', fontsize=10)

plt.grid(axis='x', linestyle='--', alpha=0.7) # Add a horizontal grid for better readability
plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()
```

```python
import re
import matplotlib.pyplot as plt
import seaborn as sns

# Function to extract the 2-character region after 5 digits
def extract_region(address):
    if pd.isna(address):
        return None
    match = re.search(r'\d{5}\s*([가-힣]{2})', str(address))
    if match:
        return match.group(1)
    return None

# Apply the function to the '주소' column to create a new '지역' column
df['지역'] = df['주소'].apply(extract_region)

# Get the value counts for the extracted regions, dropping None values
region_counts = df['지역'].dropna().value_counts()

# Filter out regions with very small counts to group them into 'Other'
threshold = 0.02 # Regions less than 2% will be grouped into 'Other'
filtered_counts = region_counts[region_counts / region_counts.sum() > threshold]
other_count = region_counts[region_counts / region_counts.sum() <= threshold].sum()

if other_count > 0:
    filtered_counts['기타'] = other_count

# Ensure the data is in ascending order for better visual (largest slices first)
# or keep it as is, depending on preference. For pie chart, largest first is common.
filtered_counts = filtered_counts.sort_values(ascending=False)

# Create the pie chart
fig, ax = plt.subplots(figsize=(10, 10)) # Adjust figure size as needed

# Define colors - using a palette for diversity
colors = sns.color_palette('pastel')[0:len(filtered_counts)]

# Explode the largest slice for emphasis (optional)
explode = [0.05 if i == 0 else 0 for i in range(len(filtered_counts))]

wedges, texts, autotexts = ax.pie(filtered_counts.values,
                                  labels=filtered_counts.index,
                                  autopct=lambda p: '{:.1f}%'.format(p) if p > 1 else '', # Only show percentages > 1%
                                  pctdistance=0.85,
                                  colors=colors,
                                  explode=explode,
                                  shadow=False, # Set to True for 3D effect
                                  startangle=90,
                                  textprops={'fontsize': 12})

# Add a title
ax.set_title('Distribution of Applications by Region', fontsize=16, pad=20)

# Equal aspect ratio ensures that pie is drawn as a circle.
ax.axis('equal')

# Adjust autopct text properties for better readability
for autotext in autotexts:
    autotext.set_color('black')
    autotext.set_fontsize(10)

plt.tight_layout()
plt.show()
```

```python
# 긴 텍스트를 짧게 정리합니다

# '타깃병원체종류' 전처리
def process_pathogen_type(text):
    if pd.isna(text): return None
    parts = str(text).split('>')
    if len(parts) >= 2: # 두 번째 단어 추출
        return parts[1].strip()
    elif len(parts) == 1: # 첫 번째 단어 사용
        return parts[0].strip()
    return None

# '신청평가유형' 전처리
def process_evaluation_type(text):
    if pd.isna(text): return None
    parts = str(text).split('>')
    if len(parts) >= 2: # 두 번째 단어 추출
        return parts[1].strip()
    elif len(parts) == 1: # 첫 번째 단어 사용
        return parts[0].strip()
    return None

# '시험종류' 전처리
def process_test_type(text):
    if pd.isna(text): return None
    parts = str(text).split('>')
    return parts[0].strip() # 첫 번째 단어만 추출

# '주소' 전처리 (우편번호 제거 후 시도명 추출)
def extract_sido(address):
    if pd.isna(address): return None
    address = str(address).strip()
    # 숫자 5자리 (우편번호) 이후 공백이 있으면 제거 후 처리
    if re.match(r'^\d{5}\s', address):
        address = re.sub(r'^\d{5}\s*', '', address, 1)

    sido_list = ['서울', '부산', '대구', '인천', '광주', '대전', '울산', '세종', '경기', '강원', '충북', '충남', '전북', '전남', '경북', '경남', '제주']
    for sido in sido_list:
        if address.startswith(sido):
            return sido
    return '기타' # 시도명 찾지 못할 경우 '기타'로 처리

import re # 정규표현식 사용을 위해 import

current_month_df['간소화_병원체'] = current_month_df['타깃병원체종류'].apply(process_pathogen_type)
current_month_df['간소화_평가유형'] = current_month_df['신청평가유형'].apply(process_evaluation_type)
current_month_df['간소화_시험종류'] = current_month_df['시험종류'].apply(process_test_type)
current_month_df['시도'] = current_month_df['주소'].apply(extract_sido)

print("--- 전처리 후 신규 데이터 (일부 컬럼) ---")
display(current_month_df[['접수일', '신청기관구분', '간소화_병원체', '간소화_평가유형', '간소화_시험종류', '시도', '진행상태']].tail(7))
```

```python
# 이번 달 현황을 숫자로 정리합니다

# 1. 총 신청 건수 (이번 달)
current_month_total_applications = len(current_month_df)

# 2. 기관구분별 건수
institution_counts = current_month_df['신청기관구분'].value_counts()

# 3. 병원체별 건수 Top 3
pathogen_counts = current_month_df['간소화_병원체'].value_counts().head(3)

# 4. 신청평가유형별 건수
evaluation_type_counts = current_month_df['간소화_평가유형'].value_counts()

# 5. 진행상태별 건수
status_counts = current_month_df['진행상태'].value_counts()

# 6. 전월 대비 증감 (2026년 3월 기준)
# 원본 df_existing에서 2026년 3월 데이터 필터링
df_existing['접수일'] = pd.to_datetime(df_existing['접수일'], errors='coerce')
prev_month_df = df_existing[
    (df_existing['접수일'].dt.year == 2026) &
    (df_existing['접수일'].dt.month == 3)
]
prev_month_applications = len(prev_month_df)

# 증감 계산
diff = current_month_total_applications - prev_month_applications
increase_decrease_direction = "증가" if diff > 0 else ("감소" if diff < 0 else "동일")
increase_decrease_value = abs(diff)
percentage_change = (diff / prev_month_applications * 100) if prev_month_applications != 0 else np.inf # 0으로 나누는 오류 방지

print("--- 2026년 4월 신청 현황 ---")
print(f"총 신청 건수: {current_month_total_applications}건")
print("\n기관구분별 건수:")
print(institution_counts)
print("\n병원체별 건수 Top 3:")
print(pathogen_counts)
print("\n신청평가유형별 건수:")
print(evaluation_type_counts)
print("\n진행상태별 건수:")
print(status_counts)
print(f"\n전월(2026년 3월) 대비 증감: {prev_month_applications}건 -> {current_month_total_applications}건 (총 {increase_decrease_value}건 {increase_decrease_direction}, {percentage_change:.2f}%) 부족한 기존 데이터로 인한 비교 불가")
```

```python
# 현황을 그래프로 자동으로 그립니다

fig, axes = plt.subplots(1, 3, figsize=(20, 7))

# 그래프 1: 기관구분별 막대그래프
sns.barplot(x=institution_counts.index, y=institution_counts.values, ax=axes[0], palette='Blues', hue=institution_counts.index, legend=False)
axes[0].set_title('2026년 4월 기관유형별 신청 현황', fontsize=14)
axes[0].set_xlabel('기관구분', fontsize=12)
axes[0].set_ylabel('신청 건수', fontsize=12)
axes[0].tick_params(axis='x', rotation=45)
for index, value in enumerate(institution_counts.values):
    axes[0].text(index, value, str(value), ha='center', va='bottom')

# 그래프 2: 병원체별 가로 막대그래프 (Top 3)
sns.barplot(x=pathogen_counts.values, y=pathogen_counts.index, ax=axes[1], palette='Greens', hue=pathogen_counts.index, legend=False)
axes[1].set_title('병원체별 신청 건수 (Top 3)', fontsize=14)
axes[1].set_xlabel('신청 건수', fontsize=12)
axes[1].set_ylabel('병원체', fontsize=12)
for index, value in enumerate(pathogen_counts.values):
    axes[1].text(value, index, str(value), ha='left', va='center')

# 그래프 3: 신청평가유형별 파이차트
# 비율이 1% 미만인 항목은 '기타'로 묶음
pie_data = evaluation_type_counts.copy()
threshold = 0.01
# 임시저장 건수가 0인 경우, 파이차트 생성을 피하기 위해 확인
if pie_data.sum() > 0:
    small_values = pie_data[pie_data / pie_data.sum() < threshold]
    if not small_values.empty:
        pie_data = pie_data[pie_data / pie_data.sum() >= threshold]
        pie_data['기타'] = small_values.sum()

labels = pie_data.index
sizes = pie_data.values

if sizes.sum() > 0: # 합계가 0인 경우 파이차트 그리지 않음
    wedges, texts, autotexts = axes[2].pie(sizes,
                                            labels=labels,
                                            autopct=lambda p: '{:.1f}%'.format(p) if p > 1 else '',
                                            pctdistance=0.85,
                                            startangle=90,
                                            colors=sns.color_palette('pastel')[0:len(labels)])
    for autotext in autotexts:
        autotext.set_color('black')
        autotext.set_fontsize(10)
    axes[2].set_title('신청평가유형 분포', fontsize=14)
    axes[2].axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
else:
    axes[2].text(0.5, 0.5, '데이터 없음', horizontalalignment='center', verticalalignment='center', fontsize=12, color='gray')
    axes[2].set_title('신청평가유형 분포', fontsize=14)
    axes[2].axis('off')

plt.tight_layout()
plt.savefig('/content/이번달_신청현황_그래프.png')
print("✅ 그래프 저장 완료: /content/이번달_신청현황_그래프.png")
plt.show()
```

### 📄 주간업무보고서 자동 업데이트 코드

아래 코드를 실행하면, 이번 주 월요일 날짜를 입력받아 주간업무보고서 (`/content/260421_주간업무보고_박성수.docx`)의 내용들을 자동 업데이트하고 새로운 파일로 저장합니다.

**자동 변경 항목:**
1.  **파일명:** 입력 날짜 기준 `YYMMDD_주간업무보고_박성수.docx` 형식
2.  **문서번호:** `KPEC-WR-PSS-XX` 형식에서 마지막 숫자 `+1`
3.  **작성일:** `YYYY.MM.DD` 형식으로 입력 날짜로 변경
4.  **날짜 필드:** 문서 내 `MM.DD` 또는 `M.D` 형식의 날짜들을 찾아 `+7일`로 변경
5.  **표 헤더 날짜 범위:** `(M.D~M.D)` 형식의 날짜 범위 텍스트를 찾아 `+7일`로 업데이트

```python
# 필요한 라이브러리 설치
!pip install python-docx

from docx import Document
from datetime import datetime, timedelta
import re
import os

# --- 헬퍼 함수 정의 ---

def parse_date_from_string(date_str, year_context=None):
    """'M.D', 'MM.DD', 'YYYY.MM.DD' 등 다양한 날짜 문자열을 datetime.date 객체로 파싱"""
    try:
        # YYYY.MM.DD 형식
        return datetime.strptime(date_str, '%Y.%m.%d').date()
    except ValueError:
        try:
            # MM.DD 형식 (년도 정보는 year_context 사용)
            if year_context:
                return datetime.strptime(f'{year_context}.{date_str}', '%Y.%m.%d').date()
            else:
                # 년도 정보가 없으면 현재 년도를 임시로 사용 (정확한 동작을 위해 year_context 권장)
                print(f"⚠️ {date_str} 날짜 파싱 시 년도 정보가 없어 현재 년도를 사용합니다.")
                return datetime.strptime(f'{datetime.now().year}.{date_str}', '%Y.%m.%d').date()
        except ValueError:
            try:
                # M.D 형식 (년도 정보는 year_context 사용)
                if year_context:
                    # MM.DD로 변환 시도 (ex: 4.5 -> 04.05)
                    month, day = map(int, date_str.split('.'))
                    return datetime.strptime(f'{year_context}.{month:02d}.{day:02d}', '%Y.%m.%d').date()
                else:
                    print(f"⚠️ {date_str} 날짜 파싱 시 년도 정보가 없어 현재 년도를 사용합니다.")
                    month, day = map(int, date_str.split('.'))
                    return datetime.strptime(f'{datetime.now().year}.{month:02d}.{day:02d}', '%Y.%m.%d').date()
            except ValueError:
                return None

def format_date_to_korean_str(date_obj, include_year=False):
    """datetime.date 객체를 'M.D' 또는 'YYYY.MM.DD' 문자열로 포맷팅"""
    if not date_obj: return None
    if include_year:
        return date_obj.strftime('%Y.%m.%d')
    else:
        return date_obj.strftime('%#m.%#d') # #m, #d는 0 제거 (Windows/Linux에서 동작 다름, %-m, %-d는 Linux)

# Linux 환경에서 #m, #d 대신 %-m, %-d 사용을 위한 처리
def format_date_no_leading_zero(date_obj):
    return f"{date_obj.month}.{date_obj.day}"


def replace_text_in_paragraph(paragraph, old_text, new_text):
    """단락 내의 텍스트를 교체합니다."""
    if old_text in paragraph.text:
        paragraph.text = paragraph.text.replace(old_text, new_text)
        return True
    return False

def replace_text_in_cell(cell, old_text, new_text):
    """셀 내의 텍스트를 교체합니다."""
    if old_text in cell.text:
        cell.text = cell.text.replace(old_text, new_text)
        return True
    return False

def update_document_content(document, input_monday_date):
    """문서 내용을 업데이트합니다."""

    # 업데이트될 날짜 및 번호 계산
    current_year = input_monday_date.year

    # 2. 문서번호: KPEC-WR-PSS-14 -> 마지막 숫자 +1
    doc_num_pattern = re.compile(r'(KPEC-WR-PSS-)(\d+)')

    # 3. 작성일: 2026.04.21 -> 입력 날짜로 변경 (YYYY.MM.DD 형식)
    old_creation_date_str_to_find = format_date_to_korean_str(input_monday_date - timedelta(days=7), include_year=True) # 지난주 월요일로 가정
    new_creation_date_str = format_date_to_korean_str(input_monday_date, include_year=True)

    # 4. 지난주 예정업무 날짜 5칸: 기존 날짜에서 +7일
    # 5. 금주 예정업무 날짜 5칸: 기존 날짜에서 +7일
    # -> MM.DD 또는 M.D 형식의 모든 날짜를 +7일 (이 부분은 문서 구조에 따라 더 정교하게 만들 수 있음)
    date_update_regex = re.compile(r'(\d{1,2}\.\d{1,2})') # MM.DD 또는 M.D 패턴 찾기

    # 6. 표 헤더 날짜 범위 텍스트도 +7일 업데이트
    # 예) 지난주 예정 업무(4.13~4.17) -> (4.20~4.24)
    # 예) 금주 예정 업무(4.20~4.24) -> (4.27~5.1)
    date_range_regex = re.compile(r'\((\d{1,2}\.\d{1,2})~(\d{1,2}\.\d{1,2})\)')


    def replace_dates_in_string(text, offset_days, year_context, replacement_type='generic'):
        if replacement_type == 'date_range':
            match = date_range_regex.search(text)
            if match:
                date1_str, date2_str = match.groups()
                date1_obj = parse_date_from_string(date1_str, year_context)
                date2_obj = parse_date_from_string(date2_str, year_context)

                if date1_obj and date2_obj:
                    new_date1_obj = date1_obj + timedelta(days=offset_days)
                    new_date2_obj = date2_obj + timedelta(days=offset_days)
                    new_date_range_str = f"({format_date_no_leading_zero(new_date1_obj)}~{format_date_no_leading_zero(new_date2_obj)})";
                    # 원래 텍스트에서 매칭된 날짜 범위만 교체
                    return text.replace(match.group(0), new_date_range_str)
            return text
        else: # generic date replacement
            new_text = text
            # 모든 MM.DD 또는 M.D 패턴을 찾아 업데이트
            for match in date_update_regex.finditer(text):
                original_date_str = match.group(1)
                date_obj = parse_date_from_string(original_date_str, year_context)
                if date_obj:
                    new_date_obj = date_obj + timedelta(days=offset_days)
                    new_date_str = format_date_no_leading_zero(new_date_obj)
                    # 정규식 패턴을 기반으로 치환하기 위해 re.sub 사용
                    # replace 대신 re.sub을 사용하여 동일한 날짜가 여러 번 나오는 경우 처리
                    new_text = re.sub(re.escape(original_date_str), new_date_str, new_text, 1) # 한 번만 치환
            return new_text


    found_doc_num = False
    # 단락 내 텍스트 업데이트
    for paragraph in document.paragraphs:
        # 문서번호 업데이트
        if doc_num_pattern.search(paragraph.text) and not found_doc_num:
            match = doc_num_pattern.search(paragraph.text)
            if match:
                old_num = int(match.group(2))
                new_num = old_num + 1
                # 수정된 부분: \g<1> 사용
                paragraph.text = doc_num_pattern.sub(f'\\g<1>{new_num}', paragraph.text)
                found_doc_num = True

        # 작성일 업데이트
        if old_creation_date_str_to_find in paragraph.text:
             paragraph.text = paragraph.text.replace(old_creation_date_str_to_find, new_creation_date_str)

        # 날짜 범위 및 일반 날짜 업데이트 (더 정교한 타겟팅 필요 시 조건 추가)
        if '지난주 예정 업무' in paragraph.text or '금주 예정 업무' in paragraph.text:
             paragraph.text = replace_dates_in_string(paragraph.text, 7, current_year, replacement_type='date_range')

        # 기타 MM.DD 형식 날짜 업데이트 (주의: 광범위하게 적용될 수 있음)
        paragraph.text = replace_dates_in_string(paragraph.text, 7, current_year, replacement_type='generic')


    # 표 내 텍스트 업데이트
    for table in document.tables:
        for row in table.rows:
            for cell in row.cells:
                # 문서번호 업데이트 (표 내에 있을 경우)
                if doc_num_pattern.search(cell.text) and not found_doc_num: # 혹시 표에 있다면
                     match = doc_num_pattern.search(cell.text)
                     if match:
                         old_num = int(match.group(2))
                         new_num = old_num + 1
                         # 수정된 부분: \g<1> 사용
                         cell.text = doc_num_pattern.sub(f'\\g<1>{new_num}', cell.text)
                         found_doc_num = True

                # 작성일 업데이트 (표 내에 있을 경우)
                if old_creation_date_str_to_find in cell.text:
                    cell.text = cell.text.replace(old_creation_date_str_to_find, new_creation_date_str)

                # 날짜 범위 및 일반 날짜 업데이트 (더 정교한 타겟팅 필요 시 조건 추가)
                if '지난주 예정 업무' in cell.text or '금주 예정 업무' in cell.text:
                    cell.text = replace_dates_in_string(cell.text, 7, current_year, replacement_type='date_range')

                # 기타 MM.DD 형식 날짜 업데이트 (주의: 광범위하게 적용될 수 있음)
                cell.text = replace_dates_in_string(cell.text, 7, current_year, replacement_type='generic')


# --- 메인 실행 로직 ---

# 1. 사용자로부터 이번 주 월요일 날짜를 입력받음
while True:
    try:
        input_date_str = input("이번 주 월요일 날짜를 YYYY-MM-DD 형식으로 입력하세요 (예: 2026-04-28): ")
        input_monday_date = datetime.strptime(input_date_str, '%Y-%m-%d').date()
        if input_monday_date.weekday() != 0: # 0은 월요일
            print("입력한 날짜는 월요일이 아닙니다. 다시 입력해주세요.")
            continue
        break
    except ValueError:
        print("잘못된 날짜 형식입니다. YYYY-MM-DD 형식으로 입력해주세요.")

print(f"입력된 이번 주 월요일: {input_monday_date}")

# 파일 경로 설정
original_docx_path = '/content/260421_주간업무보고_박성수.docx'

# 1. 파일명: 입력 날짜 기준 YYMMDD_주간업무보고_박성수.docx 형식으로 저장
new_filename = input_monday_date.strftime('%y%m%d') + '_주간업무보고_박성수.docx'
output_docx_path = os.path.join('/content', new_filename)

print(f"생성될 새 보고서 파일명: {output_docx_path}")

# 문서 로드
try:
    document = Document(original_docx_path)
    print(f"✅ 원본 문서 로드 성공: {original_docx_path}")

    # 문서 내용 업데이트
    update_document_content(document, input_monday_date)

    # 새 파일로 저장
    document.save(output_docx_path)
    print(f"✅ 업데이트된 보고서가 성공적으로 저장되었습니다: {output_docx_path}")

except FileNotFoundError:
    print(f"⚠️ 오류: 원본 파일을 찾을 수 없습니다. 경로를 확인해주세요: {original_docx_path}")
except Exception as e:
    print(f"⚠️ 문서 처리 중 오류 발생: {e}")
```

```python
import re
import os
from docx import Document
from datetime import datetime, timedelta

# --- 헬퍼 ---

def fmt(d):
    return f"{d.month}.{d.day}"

def fmt_full(d):
    return d.strftime('%Y.%m.%d')

def weekday_name(i):
    return ['월','화','수','목','금'][i]

def get_weekdays(monday):
    """월~금 5일 반환"""
    return [monday + timedelta(days=i) for i in range(5)]

def replace_in_para(para, old, new):
    """run 구조 유지하며 텍스트 교체"""
    if old not in para.text:
        return
    full = para.text.replace(old, new)
    for run in para.runs:
        run.text = ''
    if para.runs:
        para.runs[0].text = full

def process_cell(cell, replacements):
    """replacements: [(old, new), ...] 순서대로 교체"""
    for para in cell.paragraphs:
        if not para.text.strip():
            continue
        for old, new in replacements:
            if old in para.text:
                replace_in_para(para, old, new)
                break  # 셀당 하나만 매칭

# --- 메인 ---

while True:
    try:
        s = input("이번 주 월요일 날짜 입력 (예: 2026-05-04): ")
        this_monday = datetime.strptime(s, '%Y-%m-%d').date()
        if this_monday.weekday() != 0:
            print("월요일이 아닙니다. 다시 입력하세요.")
            continue
        break
    except ValueError:
        print("형식 오류. YYYY-MM-DD로 입력하세요.")

last_monday = this_monday - timedelta(days=7)

last_week = get_weekdays(last_monday)  # 지난주 월~금
this_week = get_weekdays(this_monday)  # 금주 월~금

# 헤더 범위 문자열
last_header_new = f"({fmt(last_week[0])}~{fmt(last_week[4])})"  # (4.27~5.1)
this_header_new = f"({fmt(this_week[0])}~{fmt(this_week[4])})"  # (5.4~5.8)

new_creation = fmt_full(this_monday)  # 2026.05.04

print(f"\n✅ 작성일: {new_creation}")
print(f"✅ 지난주 예정업무: {last_header_new}")
print(f"✅ 금주 예정업무:   {this_header_new}")
print(f"✅ 지난주 날짜: {[fmt(d) for d in last_week]}")
print(f"✅ 금주 날짜:   {[fmt(d) for d in this_week]}\n")

original_path = '/content/260421_주간업무보고_박성수.docx'
new_filename   = this_monday.strftime('%y%m%d') + '_주간업무보고_박성수.docx'
output_path    = os.path.join('/content', new_filename)

doc = Document(original_path)

# ── 표 순회 ──
doc_num_pattern = re.compile(r'(KPEC-WR-PSS-)(\d+)')
date_full_pattern = re.compile(r'\d{4}\.\d{2}\.\d{2}')  # YYYY.MM.DD
date_range_pattern = re.compile(r'\(\d{1,2}\.\d{1,2}~\d{1,2}\.\d{1,2}\)')
date_cell_pattern  = re.compile(r'(\d{1,2}\.\d{1,2})([월화수목금])')

for table in doc.tables:
    for row in table.rows:
        for cell in row.cells:
            text = cell.text

            # 1) 문서번호 +1
            m = doc_num_pattern.search(text)
            if m:
                new_num = int(m.group(2)) + 1
                for para in cell.paragraphs:
                    if doc_num_pattern.search(para.text):
                        replace_in_para(para,
                            m.group(0),
                            f"{m.group(1)}{new_num}")

            # 2) 작성일 (YYYY.MM.DD 형식)
            elif date_full_pattern.search(text):
                old_creation = date_full_pattern.search(text).group(0)
                replace_in_para(cell.paragraphs[0], old_creation, new_creation)

            # 3) 지난주/금주 헤더 날짜 범위
            elif '지난주 예정 업무' in text:
                old_range = date_range_pattern.search(text)
                if old_range:
                    for para in cell.paragraphs:
                        replace_in_para(para, old_range.group(0), last_header_new)

            elif '금주 예정 업무' in text:
                old_range = date_range_pattern.search(text)
                if old_range:
                    for para in cell.paragraphs:
                        replace_in_para(para, old_range.group(0), this_header_new)

            # 4) 날짜 셀 (M.D(요일) 패턴) — 지난주/금주 절대값으로 덮어쓰기
            elif date_cell_pattern.search(text):
                m2 = date_cell_pattern.search(text)
                old_str = m2.group(0) # e.g., '4.20(월)'
                day_name = m2.group(2)
                idx = '월화수목금'.index(day_name)

                # Determine col_idx to know if it's 'last week' or 'this week' column
                current_cell_col_idx = -1
                for i, c_in_row in enumerate(row.cells):
                    if c_in_row._element is cell._element:
                        current_cell_col_idx = i
                        break

                if current_cell_col_idx != -1:
                    target_date_val = None
                    if current_cell_col_idx < len(row.cells) // 2: # Last week column group
                        target_date_val = f"{fmt(last_week[idx])}({day_name})" # e.g., '5.4(월)'
                    else: # This week column group
                        target_date_val = f"{fmt(this_week[idx])}({day_name})" # e.g., '5.11(월)'

                    if target_date_val:
                        # Perform replacement directly on cell.text to ensure all occurrences are handled.
                        # This will clear existing paragraphs and create a new one with the updated text,
                        # potentially losing rich formatting within that specific cell, but ensuring the text update.
                        new_cell_text = cell.text.replace(old_str, target_date_val)
                        cell.text = new_cell_text # This automatically clears old paragraphs and adds new one

doc.save(output_path)
print(f"✅ 저장 완료: {output_path}")
```
